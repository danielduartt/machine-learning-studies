```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("/content/trees.csv")
df.head()
```

```python
sns.histplot(df['Volume'], bins = 10, color = 'r', kde = True, stat = 'density').set_title("Volume")
```

```python
df2 = pd.read_csv("/content/chicken.csv")
display(df2.head(10))
```

```python
group = df2.groupby(['feed'])['weight'].sum()
```

```python
group.plot.bar(color = 'red blue orange yellow gray green'.split())
```

```python
group_horsebean = df2.loc[df2.feed == 'horsebean']
group_horsebean
```

```python
sns.histplot(group_horsebean, kde = True)
```

```python
df2.groupby(['feed']).size()
```

```python
sns.histplot(df2.loc[df2.feed == 'casein'].weight, kde = True, color = 'r').set(title = 'casein')
```

```python
sns.histplot(df2.loc[df2.feed == 'linseed'].weight, kde = True, color = 'blue').set_title("linseed")
```

```python
sns.histplot(df2.loc[df2['feed'] == 'meatmeal'].weight, kde = True, color = 'orange').set_title("meatmeal")
```

```python
sns.histplot(df2.loc[df2['feed'] == 'soybean'].weight, kde = True, color = 'green').set_title("soybean")
```

```python
sns.histplot(df2.loc[df2['feed'] == 'sunflower'].weight, kde = True, color = 'black')
```

```python
plt.figure(figsize = (14,10))
plt.subplot(3,2,1)
sns.histplot(group_horsebean, kde = True, legend = False ).set_title("horsebean")
plt.subplot(3,2,2)
sns.histplot(df2.loc[df2.feed == 'casein'].weight, kde = True, color = 'r').set(title = 'casein')
plt.subplot(3,2,3)
sns.histplot(df2.loc[df2.feed == 'linseed'].weight, kde = True, color = 'blue').set_title("linseed")
plt.subplot(3,2,4)
sns.histplot(df2.loc[df2['feed'] == 'meatmeal'].weight, kde = True, color = 'orange').set_title("meatmeal")
plt.subplot(3,2,5)
sns.histplot(df2.loc[df2['feed'] == 'soybean'].weight, kde = True, color = 'green').set_title("soybean")
plt.subplot(3,2,6)
sns.histplot(df2.loc[df2['feed'] == 'sunflower'].weight, kde = True, color = 'black')
plt.tight_layout()
plt.show()
```
