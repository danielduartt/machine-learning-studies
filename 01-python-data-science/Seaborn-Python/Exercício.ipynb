```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/content/dados.csv", sep = ';')
df
```

```python
df['PIB'].describe()
```

```python
df['VALOREMPENHO'].describe()
```

```python
df.isnull().sum()
```

```python
df.groupby(['MUNICIPIO']).size()
```

```python
group_PIB = df.groupby(['MUNICIPIO'])['PIB'].sum()
group_PIB
```

```python
plt.figure(figsize= (40,30)) #OH GOD!
group_PIB.plot.bar()
plt.tight_layout()
plt.show()
```

```python
agrupados = df.sort_values("PIB").head(10)
agrupados = agrupados.iloc[:, 1:3]
agrupados
```

```python
agrupados.plot.bar(x = 'MUNICIPIO', y = 'PIB') #Municipio com valor do PIP
```

```python
sns.histplot(df['PIB'], kde = True, color = 'r', bins = 20).set_title("PIB")
```

```python
sns.kdeplot(df['PIB']).set_title('PIB')
```

```python
sns.histplot(df['VALOREMPENHO'], color = 'green', kde = True, bins = 10 ).set_title("VALOREMPENHO")
```

```python
sns.kdeplot(df['VALOREMPENHO'], color = 'blue').set_title("VALOREMPENHO")
```

```python
sns.boxplot(df['VALOREMPENHO'], orient = 'v', color = 'r').set_title("VALOREMPENHO")
```

```python
plt.figure(figsize = (10,8))
plt.boxplot(df['VALOREMPENHO'], patch_artist= True, showfliers = True)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.boxplot(df['PIB'], patch_artist= True, showfliers = True)
plt.tight_layout()
plt.show()
```
