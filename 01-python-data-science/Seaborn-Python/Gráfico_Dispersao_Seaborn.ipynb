```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/content/co2.csv")
df.head()
```

```python
sns.scatterplot(x = df['conc'], y = df['uptake'], hue = df['Type'])
```

```python
q = df.loc[df.Type == 'Quebec']
m = df.loc[df.Type == 'Mississippi']

plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(x = q.conc, y = q.uptake).set_title("Quebec")
plt.subplot(1,2,2)
sns.scatterplot(x = m.conc, y = m.uptake, color = 'orange').set_title("Mississippi")
plt.tight_layout()
plt.show()
```
